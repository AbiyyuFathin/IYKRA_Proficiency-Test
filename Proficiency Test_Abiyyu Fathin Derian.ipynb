### 'INCIDENT_NUMBER'
Because there are several incident number that is wrong format, I will drop them.

```python
# Get only the right format of 'INCIDENT_NUMBER'
df_IN_10 = df[(df['INCIDENT_NUMBER'].astype(str).map(len) == 10)]
df_IN_10
```

```python
# Replace 'I' into '' in 'INCIDENT_NUMBER'
df_IN_10 = df[(df['INCIDENT_NUMBER'].astype(str).map(len) == 10)]
df_IN_10['INCIDENT_NUMBER'] = df_IN_10['INCIDENT_NUMBER'].str.replace('I', '')
df_IN_10['INCIDENT_NUMBER']
```

## Modelling & Evaluation

```python
# Check info of dataset
df.info()
```

```python
# Import library
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# Encode the data
le1 = LabelEncoder()
df['OFFENSE_CODE_GROUP_N'] = le1.fit_transform(df['OFFENSE_CODE_GROUP'])
df['DISTRICT_N'] = le1.fit_transform(df['DISTRICT'])
df['DAY_OF_WEEK_N'] = le1.fit_transform(df['DAY_OF_WEEK'])
df.info()
```

```python
# Assignment of the X as the independant variable and Y as the dependant variable
X = df[['OFFENSE_CODE_GROUP_N', 'MONTH', 'DISTRICT_N', 'DAY_OF_WEEK_N']]
y = df['HOUR']
```

```python
# Divide datsaet into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size = 0.3, 
                                                    random_state = 10) 
```

```python
# Import the library of Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor

# initialize the model
model = RandomForestRegressor(random_state = 15, n_estimators = 200)

# fit the model
model.fit(X_train, y_train)
```

```python
# Predict the Y value
y_predict = model.predict(X_test)
y_predict
```

```python
# Evaluate the model
r2_score(y_test, y_predict)
```
